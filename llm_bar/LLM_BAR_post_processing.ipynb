```python
import pandas as pd
from datasets import load_dataset

LLMBar = load_dataset("princeton-nlp/LLMBar", "LLMBar")
```

```python
ground_truth = LLMBar['Natural'].to_pandas()
ground_truth.head(10)
```

```python
ground_truth.to_csv('/content/drive/MyDrive/senior_thesis/llm_bar/final_outputs/llmbar_ground_truth.csv', index=True)
```

```python
import pandas as pd
import os

folder_path = '/content/drive/MyDrive/senior_thesis/llm_bar/outputs'
dfs = {}
for filename in os.listdir(folder_path):
  if filename.endswith('.csv'):
    print(f"Processing: {filename}")

    file_path = os.path.join(folder_path, filename)
    df = pd.read_csv(file_path)
    filename = filename[:-4]
    df['id'] = df['id'].astype(int)-1

    # df = df.set_index(['id'])
    #making zero indexing
    df['pred'] = df['pred'].astype(str).str.strip()
    #A(1) is better = 1
    df['pred'] = df['pred'].map({'A': 1, 'B': 2})
    dfs[filename] = df
```

```python
dfs['llama-3-8b-it'].head(10)
```

```python
for name, df_old in dfs.items():
    df_old = df_old.merge(ground_truth, left_index = True, right_index = True, how = 'inner')  # Join on index
    df_old['correct'] = (df_old['pred'] == df_old['label']).astype(int) # Compare predictions
    dfs[name] = df_old.drop(columns=['pred', 'label'])  # Clean up by removing extra column
    # dfs[name] = df_old
```

```python
dfs['qwen-2.5-7b-it'].head(10)
```

```python
for name, df in dfs.items():
    df.to_csv(f'/content/drive/MyDrive/senior_thesis/llm_bar/final_outputs/{name}.csv', index=False)
```
