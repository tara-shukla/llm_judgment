```python
import pandas as pd
from datasets import load_dataset

ground_truth_dataset = load_dataset("ScalerLab/JudgeBench", split="claude")
```

```python
# ground_truth = LLMBar['Natural'].to_pandas()
ground_truth = ground_truth_dataset.to_pandas()
ground_truth.head()
```

```python
ground_truth['label'].unique()
```

```python
import pandas as pd

ground_truth['label'] = ground_truth['label'].str.strip().str.upper().str[0]  # Convert to uppercase if needed
ground_truth.rename(columns={'pair_id': 'id'}, inplace=True)
ground_truth.head()
```

```python
ground_truth.to_csv('/content/drive/MyDrive/senior_thesis/judgebench/judgebench_ground_truth.csv', index=True)
```

```python
import pandas as pd
import os

folder_path = '/content/drive/MyDrive/senior_thesis/judgebench/outputs'
dfs = {}
for filename in os.listdir(folder_path):
  if filename.endswith('.csv'):
    print(f"Processing: {filename}")

    file_path = os.path.join(folder_path, filename)
    df = pd.read_csv(file_path)
    filename = filename[:-4]

    df['pred'] = df['pred'].astype(str).str.strip()
    #A(1) is better = 1
    # df['pred'] = df['pred'].map({'A': 1, 'B': 2})
    dfs[filename] = df
```

```python
dfs['qwen-2.5-7b-it'].head(10)
```

```python
for name, df_old in dfs.items():
    df_old = df_old.merge(ground_truth, on = 'id')
    df_old['correct'] = (df_old['pred'] == df_old['label']).astype(int) # Compare predictions
    dfs[name] = df_old.drop(columns=['pred', 'label', 'original_id', 'response_model'])  # Clean up by removing extra column
    # dfs[name] = df_old
```

```python
dfs['qwen-2.5-7b-it'].head(10)
```

```python
for name, df in dfs.items():
    df.to_csv(f'/content/drive/MyDrive/senior_thesis/judgebench/final_outputs/{name}.csv', index=False)
```
