```python
from datasets import Dataset
from google.colab import drive
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from datasets import load_dataset
from torch.utils.data import DataLoader
from torch import cuda, tensor, no_grad
import torch
from tqdm.auto import tqdm
import json

# Output the scores as a JSON object in the following format:{{'overall': score,'cohesion': score 'syntax': score,'vocabulary': score,'phraseology': score,'grammar': score, 'conventions': score,}}
def preprocess(data):
    batch = {}
    batch['messages'] = [f"'{rubric}'\n, '{full_text}'" for full_text in data['full_text']]
    #batch['messages'] = [f"grade this essay from 1-5: {full_text}" for full_text in data['full_text']]
    batch['essay_ids'] = data['text_id_kaggle']
    return batch

def compute_metric():
    return

def collate(data):
    return {'messages': [ex['messages'] for ex in data],
            'essay_ids': [ex['essay_ids'] for ex in data]}

batch_size = 1
print('loading dataset...\n')
train_data = Dataset.from_pandas(df)
train_data = train_data.select_columns(['full_text', 'text_id_kaggle'])
train_data = train_data.map(preprocess, batched=True, batch_size=batch_size)
train_dataloader = DataLoader(train_data,
                               batch_size=batch_size, shuffle=False,
                               collate_fn=collate)
```

##LLM grading

```python

device = 'cuda' if cuda.is_available() else 'cpu'
pipe = pipeline(
    "text-generation",
    model="nvidia/Hymba-1.5B-Instruct",
    torch_dtype=torch.bfloat16,
    device_map="auto",
    return_full_text = False,
)
```

```python
epochs = 1

num_training_steps = epochs * len(train_dataloader)
progress_bar = tqdm(range(num_training_steps))

with open ('/content/drive/MyDrive/senior_thesis/outputs/hymba_raw.csv', 'w') as out_file:
    out_file.write('essay_id,pred\n')
    with no_grad():
        for _ in range(epochs):
            for i, batch in enumerate(train_dataloader):
                essay_ids = batch['essay_ids']
                for i in range(len(batch['messages'])):
                    out = pipe(batch['messages'], max_new_tokens=256, pad_token_id=pipe.tokenizer.eos_token_id)
                    out_file.write(f'{essay_ids[i]},"{out[i][0]["generated_text"]}"\n')
                    # print(f'{essay_ids[i]},"{out[i][0]["generated_text"][1]}"')
                    # print(out[i][0]["generated_text"])
                progress_bar.update()
```
