```python
import json
import re
from datasets import Dataset
import pandas as pd

# Function to clean up the JSON text
def clean_json_text(json_text):
    # Find everything between the first [ and the last ]
    match = re.search(r'\[(.*)\]', json_text, re.DOTALL)
    if match:
        json_text = '[' + match.group(1) + ']'

    # Handle incomplete JSON (if the last item is cut off)
    last_complete = json_text.rfind('},')
    if last_complete != -1:
        json_text = json_text[:last_complete+1] + ']'

    return json_text
file_name = 'drive/MyDrive/senior_thesis/MT_Bench/mt_bench.json'
# Read the file content

with open(file_name, 'r') as file:
    content = file.read()

# Clean the JSON text
cleaned_json = clean_json_text(content)

try:
    # Parse the JSON
    data = json.loads(cleaned_json)

    # Convert to DataFrame first (easier to work with)
    df = pd.DataFrame(data)

    # Create a Hugging Face Dataset
    dataset = Dataset.from_pandas(df)

    # Display info about the dataset
    print(f"Dataset created with {len(dataset)} examples")
    print("\nDataset features:")
    print(dataset.features)

    # Show a sample
    print("\nSample from the dataset:")
    print(dataset[0])

except json.JSONDecodeError as e:
    print(f"Error parsing JSON: {e}")
    print(f"Position: {e.pos}")
    print(f"Line: {e.lineno}, Column: {e.colno}")
```
